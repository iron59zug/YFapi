```
# interactive plot w/ Plotly

# libs
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import plotly.graph_objects as go

#data
stox='BABA'
sdate='2007-12-31'
```

```
# data prep
df = pdr.DataReader(stox, data_source='yahoo',start=sdate)
```

```
df.rename(columns={'High':'h', 'Low':'l', 'Open':'o', 'Close':'c', 
                   'Volume':'vol', 'Adj Close':'ac'}, inplace=True)
# show a smaller head & tail part, instead of df
dfs = df.iloc[np.r_[0:2, -2:0]] # using pd.iloc & np.r
dfht = (df.head(2)).append(df.tail(2)) # no np lib involved
print (dfht)
dfs # as shown, print() generate a more compact view, sometimes too compact
```

```
fig = go.Figure(data=[go.Candlestick(
                          x = df.index,
                          open = df.o,
                          low = df.l,
                          high = df.h,
                          close = df.ac,
                          increasing_line_color = 'green',
                          decreasing_line_color = 'red')])
fig.update_layout(title=stox,
                  yaxis_title='adj. close')
fig.show()
```
